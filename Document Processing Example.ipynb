```python
from modules.api import API
from modules.document import Document
import time
```

```python
service = API('localhost', '12345')
id = "output11"
category = "class3"
```

```python
# Upload the document for processing.
status, data = service.upload('test_docs/' + id + '.pdf', id, category)
print ("Status:", status)
print ("Message:", data["message"])
```

```python
# Polling the server to check if the document has been processed.
status, data = service.fetch(id)
while status != 200:
    status, data = service.fetch(id)
    if status != 200:
        print ("Status:", status)
        print ("Message:", data["message"])
        time.sleep(10)
    else:
        break
print ('Fetched document successfully.')
doc = Document(data)
```
